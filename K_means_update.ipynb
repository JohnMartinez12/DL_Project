```python
import numpy as np
query = np.load('feature_query_np .npy')
test = np.load('feature_test_np.npy')
test.shape
```

```python
query = np.reshape(query,(250,1000))

test = np.reshape(test,(1000,1000))
```

```python
from sklearn.cluster import KMeans
import numpy as np

X = test
kmeans = KMeans(n_clusters=len(query),init=query,random_state=0).fit(X)
#kmeans.labels_

#kmeans.predict([[0, 0, 1], [4, 4, 4]])
```

```python
pred = kmeans.predict(query)
```

```python
K = kmeans.labels_
```
